```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as img
import numpy as np
import sklearn as skl
from numpy import linalg
import pandas as pd
print('Done')
```

```python
f_names=pd.read_csv('face_names.csv')
images=[]
for i in range(len(f_names)):
    images.append(img.imread(f_names.values[i][0]))
    plt.imshow(images[i])
for i in range(7):
    print(images[i].shape)
```

```python
data=[]
for j in images:
    temp=make1D(j)
    data.append(temp)
    
```

```python
#Convert 2d image to 1d
def make1D(img):
    image1D=img.ravel().tolist()
    return image1D
```

```python
x=data[:len(f_names)-1]
y=data[len(f_names)-1]
  
```

```python
x_t=list(map(list, zip(*x)))
x_t=np.array(x_t)
xnew=list(map(list,zip(*x_t)))
temp=np.array(xnew)
```

```python
a=np.matmul(y,x_t)
b=np.matmul(temp,x_t)
#theta=np.matmul(linalg.inv(b),a)
theta=np.matmul(a,linalg.inv(b))
theta
```

```python
y_kappa=np.matmul(theta,temp)
print(y_kappa)
```

```python
finalimage = np.array(y_kappa).reshape(166,166)
```
